### HERE ARE THE EXPLAINATION OF EACH COLUMNS:

1.Glucose: It indicates the plasma glucose concentration in the person's blood, measured in milligrams per deciliter (mg/dL). Glucose levels are important in diagnosing and managing diabetes.

2.BloodPressure: This column represents the diastolic blood pressure (the pressure in the arteries when the heart is at rest) measured in millimeters of mercury (mmHg). Blood pressure is an important indicator of cardiovascular health.

3.SkinThickness: It refers to the thickness of the skinfold at the triceps site, measured in millimeters. Skinfold thickness is often used as an estimate of body fat percentage.

4.Insulin: This column represents the 2-hour serum insulin level, measured in milli-international units per milliliter (mu U/ml). Insulin is a hormone that regulates blood sugar levels, and low or high insulin levels can indicate potential issues with glucose metabolism.

5.BMI (Body Mass Index): It is a numerical value calculated using a person's weight and height. It is obtained by dividing the weight in kilograms by the square of the height in meters. BMI provides an estimate of body fat and is commonly used to assess overweight and obesity.

6.DiabetesPedigreeFunction: This column represents a measure of the genetic influence of diabetes based on family history. It provides an estimation of the likelihood of developing diabetes based on genetics.

7.Age: It indicates the age of the person in years.

8.Outcome: This column serves as the target variable or dependent variable. It represents whether a person has diabetes (1) or does not have diabetes (0). It is used for classification tasks, where the goal is to predict the presence or absence of diabetes based on the other features in the dataset.

9.Gender: This column represents the gender of the person if he/she is male/female.

## DATA CLEANING

```python
mean_glucose = data['Glucose'].mean()
data['Glucose'] = data['Glucose'].where(data['Glucose'] != 0, mean_glucose)
mean_BloodPressure = data['BloodPressure'].mean()
data['BloodPressure'] = data['BloodPressure'].where(data['BloodPressure'] != 0, mean_BloodPressure)
mean_SkinThickness = data['SkinThickness'].mean()
data['SkinThickness'] = data['SkinThickness'].where(data['SkinThickness'] != 0, mean_SkinThickness)
mean_Insulin = data['Insulin'].mean()
data['Insulin'] = data['Insulin'].where(data['Insulin'] != 0, mean_Insulin)
mean_BMI = data['BMI'].mean()
data['BMI'] = data['BMI'].where(data['BMI'] != 0, mean_SkinThickness)
mean_DiabetesPedigreeFunction = data['DiabetesPedigreeFunction'].mean()
data['DiabetesPedigreeFunction'] = data['DiabetesPedigreeFunction'].where(data['DiabetesPedigreeFunction'] != 0, mean_DiabetesPedigreeFunction)
mean_Age = data['Age'].mean()
data['Age'] = data['Age'].where(data['Age'] != 0, mean_Age)

data.to_csv('dataset_update.csv', index=False)
```

## Distributions

```python
from scipy.stats import shapiro, boxcox
import matplotlib.pyplot as plt

column = 'Glucose'
mean = df[column].mean()
std = df[column].std()

q1 = df[column].quantile(0.25)
q3 = df[column].quantile(0.75)
iqr = q3 - q1

plt.hist(df[column])
plt.xlabel(column)
plt.ylabel('Count')
plt.title('Distribution of {}'.format(column))
plt.show()

normality_test = shapiro(df[column])
print('p-value for normality test:', normality_test[1])

df[column], _ = boxcox(df[column] + 1)  

mean_transformed = df[column].mean()
std_transformed = df[column].std()

q1_transformed = df[column].quantile(0.25)
q3_transformed = df[column].quantile(0.75)
iqr_transformed = q3_transformed - q1_transformed

plt.hist(df[column])
plt.xlabel(column)
plt.ylabel('Count')
plt.title('Distribution of transformed {}'.format(column))
plt.show()

normality_test_transformed = shapiro(df[column])
print('p-value for normality test after Box-Cox transformation:', normality_test_transformed[1])
```

## Hypothesis Testing

```python
from scipy.stats import ttest_ind

male_bmi = data[data['Gender'] == 'Male']['BMI']
female_bmi = data[data['Gender'] == 'Female']['BMI']
t_stat_bmi, p_value_bmi = ttest_ind(male_bmi, female_bmi)
print(f"Hypothesis Test for BMI Difference: p-value = {p_value_bmi}")
print(f"Hypothesis Test for BMI Difference: t-statistic = {t_stat_bmi}")
```
